```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('__file__').parent.resolve()))
from aem_csv_to_xyz import model_csv_to_xyz
```

```python
# =============================================================================
# Configuration
# =============================================================================

csv_file    = '/path/to/your/inversion_model_file.csv'
output_file = None   # None = same directory and base name, .xyz extension

# Active unit system. Currently informational only — controls which column
# group gets canonical names via column_map below.
# TODO: derive from CRS / EPSG code automatically.
units = 'm'

# Map original column names to canonical libaarhusxyz names.
# For layer groups (e.g. DEP_TOP_M[0..N]), supply only the prefix as the key.
# Columns absent from this dict pass through lowercased but otherwise
# unchanged (e.g. RHO_I[N], RHO_I_STD[N], SIGMA_I[N] are detected as layer
# data automatically by libaarhusxyz via their [N] notation).
# Only meter-unit columns are mapped here — ft-unit columns remain in the
# file but will not be used by libaarhusxyz for coordinate / depth calculations.
#
# Example mapping for a SkyTEM SCI/LCI model file with meter-unit columns:
column_map = {
    # Scalar columns
    'Line':         'line_no',   # flight-line identifier
    'East_UTM_M':   'x',
    'North_UTM_M':  'y',
    'DEM_M':        'elevation',
    'ALT_M':        'alt',       # sensor altitude above ground
    'DOI_UPPER_M':  'doi_conservative',
    'DOI_LOWER_M':  'doi_standard',
    # Layer group prefixes (applied to all [N] columns in each group)
    'DEP_TOP_M':    'dep_top',
    'DEP_BOT_M':    'dep_bot',
    'THK_M':        'thk',
}

# Columns to remove entirely from the output (case-insensitive).
# Leave empty to keep all columns.
drop_columns = []
```

```python
out = model_csv_to_xyz(
    csv_file,
    output_file=output_file,
    units=units,
    column_map=column_map,
    drop_columns=drop_columns,
)
print(f'Output: {out}')
```
